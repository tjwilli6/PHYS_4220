```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler1(dt=0.01,g=-9.8,v0=10):
    t = np.arange(0,10,dt)
    #v = np.zeros( (t.size) )
    v = np.zeros_like(t)
    v[0] = v0
    for i in range(1,t.size):
        v[i] = v[i-1] + g * dt
    return t,v
```

```python
t,v = euler1()
```

```python
plt.plot(t,v)
plt.plot(t,-9.8*t+10)
```

```python
for dt in (2,0.1,0.05,0.01):
    t,v = euler1()
    plt.plot(t,v)
plt.plot(t,-9.8*t+10)
```

```python
def euler2(v0=0,dt=.01):
    t = np.arange(0,5,dt)
    v = np.zeros_like(t)
    v[0] = v0
    
    for i in range(1,t.size):
        v[i] = v[i-1] + ( 1 - v[i-1] ) * dt
    return t,v
```

```python
t,v = euler2()
t1,v1 = euler2(v0=2)
```

```python
plt.plot(t,v)
plt.plot(t1,v1)
```
