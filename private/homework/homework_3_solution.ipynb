```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Video
import os
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
def racc(r):
    return 1 / r**3 - 2*r

def get_motion(r0,vr0,tstop=10,dt=0.01):
    t = np.arange(0,tstop,dt)
    r = np.zeros_like(t)
    rdot = np.zeros_like(t)
    phi = np.zeros_like(t)
    
    r[0] = r0
    rdot[0] = vr0
    
    for i in range(1,t.size):
        rdot[i] = rdot[i-1] + racc(r[i-1]) * dt
        r[i] = r[i-1] + rdot[i] * dt
        phi[i] = phi[i-1] + dt / r[i]**2
    return t,r,phi


def plot_xy(**kwds):
    x = r * np.cos( phi )
    y = r * np.sin( phi )
    kwds.setdefault('color','blue')
    plt.plot(x,y,**kwds)
    plt.gca().set_aspect('equal')
    plt.xlabel('X')
    plt.ylabel('Y')
```

```python
r0 = (1/2)**(1/4)
t,r,phi = get_motion(r0,0,dt=1e-3)
```
