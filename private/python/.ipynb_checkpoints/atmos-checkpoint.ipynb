```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.style.use("/Users/tjwilli/presentation.mplstyle")
```

```python
!head atmos.csv
```

```python
std_model = np.genfromtxt("atmos.csv",delimiter=",",usecols=range(1,7),skip_header=1)
data = np.genfromtxt("/Users/tjwilli/Desktop/atmprof21.dat")
```

```python
plt.figure(figsize=(8,6))
plt.plot(std_model[:,0]/1e3,std_model[:,4]/1e3)
plt.plot(data[:,0],data[:,1])
plt.ylabel(r'$\rho$')
plt.xlabel('y')
```

```python
def iso_model(h,p0,h0):
    return p0 * np.exp(-h/h0)
def adiab_model(h,p0,a,T0,alpha):
    return p0 * (1-a*h/T0)**alpha
```

```python
popt,pcov = curve_fit(iso_model,data[:,0],data[:,1])
popt1,pcov1 = curve_fit(adiab_model,data[:,0],data[:,1],p0=[1e-3,6e-3,270,2.5])
```

```python
plt.figure(figsize=(10,4))
plt.scatter(data[:,0],data[:,1],label='data',color='red')
xplot = np.linspace(data[:,0].min(),data[:,0].max(),1000)
yplot = iso_model(xplot,*popt)
yplot1 = adiab_model(xplot,*popt1)
plt.plot(xplot,yplot,label='isothermal',color='green')
plt.plot(xplot,yplot1,label='adiabatic',color='blue')
#plt.ylim(-data[:,1].max()/10,data[:,1].max()*1.2)
plt.ylim(1e-12)
plt.yscale('log')
plt.legend()
plt.xlabel('altitude [km]')
plt.ylabel('density [g/cm3]')
```
