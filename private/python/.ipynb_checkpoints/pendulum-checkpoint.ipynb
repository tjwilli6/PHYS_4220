# Simple pendulum
### $\frac{d^2\theta}{dt^2} = -\theta$
#### Solved with standard Euler

#### Solve

```python
dt = 0.05
t = np.arange(0,8 * 2 * np.pi,dt)

#Arrays to collect theta and w
theta = np.zeros_like(t)
omega = np.zeros_like(t)

#Start with theta at 1 degree
theta[0] = np.deg2rad(1)

for i in range(1,t.size):
    omega[i] = omega[i-1] - theta[i-1] * dt
    theta[i] = theta[i-1] + omega[i-1] * dt
```

#### Plot

```python
plt.plot(t,theta,c='blue')
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\theta$')
plt.title('Amplitude vs time')
```

### Check on Energy conservation
### $E=1 + \frac{1}{2}\omega^2 -\cos{\theta}$

```python
E = 0.5*omega**2 + ( 1 - np.cos(theta) )
plt.plot(t,(E-E[0])/E[0],c='blue')
plt.plot(t,np.exp(t*dt) - 1,c='k',ls='--')
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\frac{E}{E_0}$')
plt.title('Energy vs time')
```

## Projectile motion energy?

```python
dt = 0.05
t = np.arange(0,10,dt)

#velocities
vx = np.zeros_like(t)
vy = np.zeros_like(t)
#positions
x = np.zeros_like(t)
y = np.zeros_like(t)

theta = np.deg2rad( 45 )
vx[0] = np.cos( theta )
vy[0] = np.sin( theta )

for i in range(1,t.size):
    vx[i] = vx[i-1]
    x[i] = x[i-1] + vx[i] * dt
        
    vy[i] = vy[i-1] - dt
    y[i] = y[i-1] + vy[i-1] * dt

E = 0.5*(vx**2+vy**2) + y
plt.plot(t,(E-E[0])/E[0],c='blue')
plt.plot(t,0.5*dt*t/E[0],c='k',ls='--')
plt.xlabel(r'$\frac{t}{t_0}$')
plt.ylabel(r'$\frac{E}{E_0}$')
```

# Damped pendulum
### $\frac{d^2\theta}{dt^2} = -\theta - Q\frac{d\theta}{dt}$
#### Solved with Euler-Cromer

```python
def damped_pend(theta0,Q,dt=0.01):
    t = np.arange(0,8 * 2 * np.pi,dt)
    #Arrays to collect theta and w
    theta = np.zeros_like(t)
    omega = np.zeros_like(t)

    #Start with theta at 1 degree
    theta[0] = np.deg2rad(theta0)

    for i in range(1,t.size):
        omega[i] = omega[i-1] - theta[i-1] * dt - Q * omega[i-1] * dt
        theta[i] = theta[i-1] + omega[i] * dt

    return t,theta
```

```python
qvals = { 0.2:'underdamped', 2:'critically damped', 4:'overdamped'}
for key,val in qvals.items():
    t,theta = damped_pend(1,key)
    plt.plot(t,theta,label=val)
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\theta$')
plt.title('Amplitude vs time')
plt.legend()
```

### Damped & driven pendulum:
### $\frac{d^2\theta}{dt^2}=-\theta-Q\frac{d\theta}{dt}+A\sin{\Omega_D t}$

```python
def driven_pend(theta0,Q,wd=1,A=1,dt=0.01):
    t = np.arange(0,20 * 2 * np.pi,dt)
    #Arrays to collect theta and w
    theta = np.zeros_like(t)
    omega = np.zeros_like(t)

    #Start with theta at 1 degree
    theta[0] = np.deg2rad(theta0)

    for i in range(1,t.size):
        omega[i] = omega[i-1] - theta[i-1] * dt - Q * omega[i-1] * dt + A * np.sin( wd * t[i] ) * dt
        theta[i] = theta[i-1] + omega[i] * dt

    return t,omega,theta
```

```python
Q = 0.1
wd = 1.6
A = .1
t,omega,theta = driven_pend(1,Q,wd=wd,A=A,dt=0.01)
plt.plot(t,theta)
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\theta$')
plt.title('Amplitude vs time')
```

### Non-linearities: $\sin{\theta}\neq\theta$
### $\frac{d^2\theta}{dt^2}=-\sin{\theta}$

```python
theta = np.linspace(0,np.pi,1000)
plt.plot(np.rad2deg(theta),theta,label=r'$\theta$')
plt.plot(np.rad2deg(theta),np.sin(theta),label=r'$\sin(\theta)$')
plt.xlabel(r'$\theta$')
plt.legend()
```

```python
def nonlinear_pend(theta0,dt=0.01,num_cycles=10):
    t = np.arange(0,num_cycles * 2 * np.pi,dt)
    #Arrays to collect theta and w
    theta = np.zeros_like(t)
    omega = np.zeros_like(t)

    #Start with theta at 1 degree
    theta[0] = np.deg2rad(theta0)

    for i in range(1,t.size):
        omega[i] = omega[i-1] - np.sin( theta[i-1] ) * dt
        theta[i] = theta[i-1] + omega[i] * dt

    return t,omega,theta
```

```python
theta0 = 30
t,w,theta = nonlinear_pend(theta0)
plt.plot(t,theta,c='blue',label='numerical solution')
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\theta$')
plt.title('Amplitude vs time')
plt.legend(loc='lower left')
```

```python
theta0 = 60
t,w,theta = nonlinear_pend(theta0)
plt.plot(t,theta,c='blue',label='numerical solution')
plt.xlabel(r'$\Omega t$')
plt.ylabel(r'$\theta$')
plt.title('Amplitude vs time')
plt.legend(loc='lower left')
plt.plot(t,np.deg2rad(theta0)*np.cos(t),c='k',ls='--',label=r'$\theta_0\cos{(t)}$')
```

### Driven, damped, nonlinear pendulum
### $\frac{d^2\theta}{dt^2}=-\sin{\theta} - Q\frac{d\theta}{dt} + A\sin{\Omega_D t}$

```python
def physical_pend(theta0,Q=.1,wd=1,A=1,dt=0.01,num_cycles=10):
    t = np.arange(0,num_cycles * 2 * np.pi,dt)
    #Arrays to collect theta and w
    theta = np.zeros_like(t)
    omega = np.zeros_like(t)
    #Start with theta at 1 degree
    theta[0] = np.deg2rad(theta0)

    for i in range(1,t.size):
        omega[i] = omega[i-1] - np.sin( theta[i-1] )  * dt - Q * omega[i-1] * dt + A * np.sin( wd * t[i] ) * dt
        theta[i] = theta[i-1] + omega[i] * dt
        
        #Keep angle within -pi,pi
        if ( theta[i] < -np.pi ):
            theta[i] += 2 * np.pi
        elif ( theta[i] >= np.pi ):
            theta[i] -= 2 * np.pi

    return t,omega,theta
```

### No driving force

```python
t0 = 30
wd = 2/3
A = 0
Q = 0.2
t,w,theta = physical_pend(t0,Q=Q,A=A,wd=wd,num_cycles=30,dt=0.01)
```

```python
def autocorr(array,offset):
    n = int(offset)
    norm = np.dot(array,array) / ( array.size ) * (array.size-n)
    if n == 0:
        a1 = array
        a2 = array
    else:
        a1 = array[:-n] - array.mean()
        a2 = array[n:] - array.mean()
    return np.dot(a1,a2) / norm
```

```python
t0 = 20
wd = 2/3
A = 1.3
Q = 0.1
t,w,theta = physical_pend(t0,Q=Q,A=A,wd=wd,num_cycles=30,dt=0.01)
```

```python
toff = np.arange(0,t.size,10)
acf = [ autocorr(theta,ioff) for ioff in toff ]
```

```python
plt.figure(figsize=(10,4))
plt.plot(toff,acf,c='red')
plt.axhline(0,c='k')
plt.ylim(-1.1,1.1)
```
