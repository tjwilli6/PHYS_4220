```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
plt.style.use("/Users/tjwilli/presentation.mplstyle")
```

```python
def deriv(tn,yn):
    """Dimensionless derivative"""
    return -yn

#Time step
dt = 0.01 #in dimensionless units (0.1*Tau)

#Initial value
t0 = 0
y0 = 5

#The time value when we want to stop the loop
tstop = 5 #in dimensionless units (5*Tau)
```

```python
time = np.arange(t0,tstop,dt)
y = np.zeros_like(time)
yactual = np.zeros_like(time)

time[0] = t0
y[0] = y0

for i in range(1,time.size):
    tprev = time[i-1]
    yprev = y[i-1]
    
    #Calculate derivative
    dydtprev = deriv(tprev,yprev)
    yn = yprev + dydtprev * dt
    y[i] = yn
```

```python
plt.figure(figsize=(6,4))
plt.plot(time,y,label='approx')
plt.plot(time,y0*np.exp(-time),label='actual')
plt.title("dt={}".format(dt))
plt.xlabel(r'$\frac{t}{\tau}$')
plt.ylabel('N')
plt.legend();
```

## At what time $t$ will $y(t)=0.25y_0$?

```python
target = 0.25
yfrac = y / y0
yminus = np.abs( yfrac - target )
index = yminus.argmin()
t_target = time[index]
y_target = yfrac[index]
print( 'y(t)={:.2f} at t={:.2f}'.format(y_target,t_target) )
```
