## Normalization:
### $r_0=\frac{1}{GM}\left(\frac{\ell}{\mu}\right)^2 \leftarrow$Equilibrium orbit
### $t_0=\left(\frac{r_0^3}{GM}\right)^{1/2}\leftarrow$1/(frequency of oscillations about $r_0$)
### $\ddot{r}=\frac{1}{r^3}-\frac{1}{r^2}$

```python
def orbit(vr0=0,r0=1,dt=0.01,tstop=10):
    t = np.arange(0, tstop * 2 * np.pi, dt)
    r,phi,rdot,phidot = ( np.zeros_like( t ) for i in range(4) )
    
    r[0] = r0
    rdot[0] = vr0
    phi[0] = 0
    
    for i in range(1,t.size):
        phidot[i] = 1 / r[i-1]**2
        phi[i] = phi[i-1] + phidot[i] * dt
        rdot[i] = rdot[i-1] + ( 1 / r[i-1]**3 - 1 / r[i-1]**2 ) * dt
        r[i] = r[i-1] + rdot[i] * dt
        
    result = {
        't':t,
        'r':r,
        'rdot':rdot,
        'phi':phi,
        'phidot':phidot
    }
    
    return result
```

## $r_{initial}=r_0$, $\dot{r}_0=0$, $E = U_{eff}(r_0) < 0$

```python
res = orbit(r0=1,vr0=5,tstop=20000)
x = res['r'] * np.cos(res['phi'])
y = res['r'] * np.sin(res['phi'])
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
```

```python
plt.plot(res['t'],res['rdot'])
plt.axhline(0)
```

### $\vec{r_1}=\vec{R}-\frac{m_2}{m_1+m_2}\vec{r}=-\frac{u}{1+u}\vec{r}; u=\frac{m2}{m1}$
### $\vec{r_2}=\frac{1-u}{1+u}\vec{r}$

```python
res = orbit(r0=1,vr0=.4,tstop=3)
x = res['r'] * np.cos(res['phi'])
y = res['r'] * np.sin(res['phi'])
r = 0.9
x1 = -r * x / (1+r)
y1 = -r * y / (1+r)
x2 = (1-r) * x / (1+r)
y2 = (1-r) * y / (1+r)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
```
