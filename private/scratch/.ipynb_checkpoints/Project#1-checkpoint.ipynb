```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Two Balls where r = m2/m1.

```python
def get_motion_2(r,tstop = 10, dt=.001):
    #creates inital array for the variables in use
    t = np.arange(0, tstop*np.sqrt(2),dt)
    v1 = np.zeros_like(t)
    v2 = np.zeros_like(t)
    y1 = np.zeros_like(t)
    y2 = np.zeros_like(t)
    
    #creates lists used for the Poincar ÃÅe section
    Py2 = []
    Pv2 = []
    
    #initial heights for the balls
    y1[0] = .5
    y2[0] = 1
    
    #loops over the time range
    for i in range(1,t.size):
        #increments thed velocities and heights of the balls
        v1[i] = v1[i-1] - dt
        v2[i] = v2[i-1] - dt
        y1[i] = y1[i-1] + v1[i] * dt
        y2[i] = y2[i-1] + v2[i] * dt
        #check to see if the bottome ball hit the ground
        if y1[i] <= 0:
            v1[i] = -v1[i-1]
            y1[i] = y1[i-1] + v1[i] *dt
            #records the Poincare information
            Py2.append(y2[i])
            Pv2.append(v2[i])
        #checks to see if the bottom ball hit the top ball
        if y1[i] >= y2[i]:
            #reaction onto the bottom ball
            v1[i] = ((1-r)/(r+1))*v1[i-1] + (2*r/(r+1))*v2[i-1]
            y1[i] = y1[i-1] + v1[i]*dt
            #reaction onto the top ball
            v2[i] = (2/(r+1))*v1[i-1]+((r-1)/(r+1))*v2[i-1]
            y2[i] = y2[i-1] + v2[i]*dt
    return t, v1, y1, v2, y2, Py2, Pv2
```

```python
#calls function to get values for r>1
r=2
t, v1, y1, v2, y2, Py2, Pv2 = get_motion_2(r)

#plots the trajectory
plt.figure(0)
plt.plot(t,y1)
plt.plot(t,y2)
plt.title("Trajectory for r>1")
plt.xlabel('t')
plt.ylabel("y")

#plot kinectic, potential, total energy
Pot = (1/r+1)*(y1 + r*y2)
Kin = (1/r+1)*(.5*(v1**2) + .5*r*(v2**2))
Tot = Pot + Kin
plt.figure(1)
plt.plot(t,Pot, label = "Potential")
plt.plot(t,Kin, label = "Kinetic")
plt.plot(t,Tot, label = "Total")
plt.title("Potential, Kinetic, and Total energy for r>1")
plt.legend(loc = 'lower left')

#plots the trajectory of each ball in phase space
plt.figure(2)
plt.plot(y1,v1)
plt.plot(y2,v2)
plt.title("Phase Space for r>1")
plt.xlabel('y')
plt.ylabel("v")

#plot the Poincare section for r>1
plt.figure(3)
plt.scatter(Py2, Pv2)
plt.xlabel(r'$\frac{t}{t_0}$')
plt.ylabel(r'$\frac{y}{y_0}$')
plt.title("Poincare section for r>1")
plt.xlabel('y2')
plt.ylabel("v2")
```
