2-bodies

```python
def orbit(vr0 = 0, r0 = 1, dt = .01, tstop = 10):
    
    # Delcare Array
    t = np.arange(0, tstop*2*np.pi, dt)
    
    # Create Arrays
    r = np.zeros_like(t)
    phi = np.zeros_like(t)
    rdot = np.zeros_like(t)
    phidot = np.zeros_like(t)
    
    # Initial Conditions number 1
    r[0] = r0
    rdot[0] = vr0
    phi[0] = 0
    
    for i in range(1, t.size):
        phidot[i] = 1/r[i-1]**2
        phi[i] = phi[i-1] + phidot[i]*dt
        rdot[i] = rdot[i-1]+(1/r[i-1]**3 -1/r[i-1])*dt
        r[i] = r[i-1] + rdot[i]*dt
        
    result = {
        't':t,
        'r':r,
        'rdot':rdot,
        'phi':phi,
        'phidot': phidot,
    }
    return result
```

```python
res = orbit(r0 = 1, vr0 = 0, tstop = 10)
x = res['r'] *np.cos(res['phi'])
y = res['r'] *np.sin(res['phi'])
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
```

3-bodies

```python
def threebody(x0:list, y0:list, vx0:list, vy0:list, m:list, dt:float=.01,tstop:float=10):
    t=np.arange(0, tstop,dt)
    
    vx1 = np.zeros_like(t)
    vy1 = np.zeros_like(t)
    vx2 = np.zeros_like(t)
    vy2 = np.zeros_like(t)
    vx3 = np.zeros_like(t)
    vy3 = np.zeros_like(t)
    x1 = np.zeros_like(t)
    y1 = np.zeros_like(t)
    x2 = np.zeros_like(t)
    y2 = np.zeros_like(t)
    x3 = np.zeros_like(t)
    y3 = np.zeros_like(t)
    
    # Normalize masses
    m = m / np.sum(m)
    
    # Initial Conditions
    vx1[0] = vx0[0]
    vy1[0] = vy0[0]
    vx2[0] = vx0[0]
    vy2[0] = vy0[0]
    vx3[0] = vx0[0]
    vy3[0] = vy0[0]
    # Positions
    x1[0] = x0[0]
    y1[0] = y0[0]
    x2[0] = x0[0]
    y2[0] = y0[0]
    x3[0] = x0[0]
    y3[0] = y0[0]
    
    for i in range(1,t.size):
        
        dvx1 = (
            -m[1]*(x1[i-1] - x2[i-1]) / np.hypot(x1[i-1] - x2[i-1], y1[i-1] -y2[i-1])**3
            -m[2] * (x1[i-1] - x3[i-1]) / np.hypot(x1[i-1] - x3[i-1], y1[i-1] - y3[i-1])**3
        )
        
        dvy1 = (
            -m[1]*(y1[i-1] - y2[i-1]) / np.hypot(x1[i-1] - x2[i-1], y1[i-1] -y2[i-1])**3
            -m[2] * (y1[i-1] - y3[i-1]) / np.hypot(x1[i-1] - x3[i-1], y1[i-1] - y3[i-1])**3
        
        )
        
        dvx2 = (
            -m[1]*(x2[i-1] - x1[i-1]) / np.hypot(x2[i-1] - x1[i-1], y2[i-1] -y1[i-1])**3
            -m[2] * (x2[i-1] - x3[i-1]) / np.hypot(x2[i-1] - x3[i-1], y2[i-1] - y3[i-1])**3
        )
        
        dvy2 = (
            -m[1]*(y2[i-1] - y1[i-1]) / np.hypot(x2[i-1] - x1[i-1], y2[i-1] -y1[i-1])**3
            -m[2] * (y2[i-1] - y3[i-1]) / np.hypot(x2[i-1] - x3[i-1], y2[i-1] - y3[i-1])**3
        
        )
        
        dvx3 = (
            -m[1]*(x3[i-1] - x1[i-1]) / np.hypot(x3[i-1] - x1[i-1], y3[i-1] -y1[i-1])**3
            -m[2] * (x3[i-1] - x2[i-1]) / np.hypot(x3[i-1] - x2[i-1], y3[i-1] - y2[i-1])**3
        )
        
        dvy3 = (
            -m[1]*(y3[i-1] - y1[i-1]) / np.hypot(x3[i-1] - x1[i-1], y3[i-1] -y1[i-1])**3
            -m[2] * (y3[i-1] - y2[i-1]) / np.hypot(x3[i-1] - x2[i-1], y3[i-1] - y2[i-1])**3
        
        )
        
        # Calculate veloctiry at t = i*dt
        vx1[i] = vx1[i-1] + dvx1 * dt
        vx2[i] = vx2[i-1] + dvx2 * dt
        vx3[i] = vx3[i-1] + dvx3 * dt
        
        # Euler-Cramer
        x1[i] = x1[i-1] + vx1[i] * dt
        x2[i] = x2[i-1] + vx2[i] * dt
        x3[i] = x3[i-1] + vx3[i] * dt
        
        vy1[i] = vy1[i-1] + dvy1 * dt
        vy2[i] = vy2[i-1] + dvy2 * dt
        vy3[i] = vy3[i-1] + dvy3 * dt
        
        y1[i] = y1[i-1] + vy1[i] * dt
        y2[i] = y2[i-1] + vy2[i] * dt
        y3[i] = y3[i-1] + vy3[i] * dt
        
    x = [x1, x2, x3]
    y = [y1, y2, y3]
    vx = [vx1, vx2, vx3]
    vy = [vy1, vy2, vy3]
        
    return vx, vy, x, yv
```
