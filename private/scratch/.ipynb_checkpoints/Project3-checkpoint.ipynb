```python
#Import Packages
import numpy as np
import matplotlib.pyplot as plt
import math
import time
import array
```

```python
# Potential energy
def potential(x,d,V0,x0):
    v = np.zeros_like(x)
    for i in range(x.size):
        if x[i] >= x0 and x[i] <=x0+d:
            v[i] = V0
    return v
```

```python
# Initial Gaussian moving with momentum
def psi_initial(x, x0, sigma, k=350):
    return(((1/(np.pi*sigma**2))**(1/4))*
           (np.exp((-1/2)*((x -x0)**2/sigma**2)))*
           (np.exp(1j*k*(x - x0))) )
```

```python
# Initital Conditions
x0 = 0.3
sigma = 0.05
k =350

# Time and position steps
dx = 1e-3
dt = 1e-7

# Grid Parameters
x = np.arange(0,1.5,dx)
t = np.arange(0, 2e-3, dt)

# Potential Barrier
x01 = 0.6
d = 5e-3
E = k**2 / 2 
#V0 = [2*E]
V0 = [0, .2*E, .4*E, .6*E, .8*E, E, 1.4*E, 1.6*E, 1.8*E, 2*E]

```

```python
# Normalizing the psi equation
psi0 = psi_initial(x,x0,sigma)
conjugate = np.conj(psi0)*psi0
plt.plot(x, conjugate.real)
psi_t = [[0 for x in range(len(psi0))] for y in range(len(t))]
psi_t[0] = psi0
```

```python
# Checking Normalization
integral = np.trapz(np.conj(psi0)*psi0,x)
print('The integral is', integral)
```

```python
# Getting matrix C
def C_matrix(v,dx,dt):
    Hdiag = np.diag(-1j * (1/dx**2 + v), 0)
    offdiag_elems = np.ones_like(x)[:-1] * 1j / 2 / dx**2
    Hdiaglow = np.diag(offdiag_elems, k=-1)
    Hdiaghi = np.diag(offdiag_elems, k =1)
    Hoffdiag = Hdiaglow + Hdiaghi
    H = Hdiag + Hoffdiag
    
    iden = np.identity(x.size)
    A = iden - dt / 2 * H
    B = iden + dt / 2 * H
    
    Ainv = np.linalg.inv(A)
    C = np.matmul(Ainv, B)
    return C
```

```python
for i in range(0, len(V0)):
    # Getting the potential energy
    V = potential(x, d, V0[i], x01)
    # Printing matrix C
    C = C_matrix(V, dx, dt)
    # Psi equation many times
    psi_t[0] = psi0
    for timeHere in range(1,len(t)):
        psi_t[timeHere] = np.dot(C, psi_t[timeHere-1])
    prob = np.conj(psi_t[-1]) * psi_t[-1]
    totalprob = np.trapz(prob,x)
    leftX = []
    leftProb = []
    for z in range(len(x)):
        if x[z]<x01:
            leftX.append(x[z])
            leftProb.append(prob[z])
    trans = np.trapz(leftProb,leftX)
    print("The probablility of transmission for V0 = " +  str(V0[i]) + " is " + str(trans.real))
    reflect = np.trapz(prob[x>x01+d],x[x>x01+d])
    print("The probablility of reflection for V0 = " +  str(V0[i]) + " is " + str(reflect.real))
    plt.plot(x, np.real(prob), label = 't=(:.2E)'.format(timeHere))
    plt.figure()
plt.plot()
```

```python
for n in range(0, len(psi_t), 2000):
    timeNow = dt *n 
    prob = np.conj(psi_t[n]) * psi_t[n]
    plt.plot(x, np.real(prob))
```
