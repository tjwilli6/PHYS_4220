```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from math import sqrt
```

```python
#r<1
dt = .001
tstop = 50
time = np.arange(0,tstop,dt)
m1 = 2
m2 = 1
M = m1+m2
g = 9.8
y1 = np.zeros_like(time)
y2 = np.zeros_like(time)
y1[0] = 20
y2[0] = 50 #y2 needs to be top ball
ke1 = np.zeros_like(time)
pe1 = np.zeros_like(time)
ke1[0] = 0
pe1[0] = m1*g*y1[0]
ke2 = np.zeros_like(time)
pe2 = np.zeros_like(time)
ke2[0] = 0
pe2[0] = m2*g*y2[0]
v1 = np.zeros_like(time)
v1[0] = 0
v2 = np.zeros_like(time)
v2[0] = 0
totalEnergy1 = np.zeros_like(time)
totalEnergy1[0] = ke1[0]+pe1[0]
totalEnergy2 = np.zeros_like(time)
totalEnergy2[0] = ke2[0]+pe2[0]
totalEnergy = np.zeros_like(time)
totalEnergy[0] = totalEnergy1[0]+totalEnergy2[0]
flag = False
r = m2/m1
pointv = [0]
pointy = [0]

#t_last = -sqrt(2*h0/g) # time we would have launched to get to h0 at t=0
v1max = sqrt(2 * y1[0] * m1* g)
v2max = sqrt(2 * y2[0] * m2* g)
for i in range (1,len(time)):
    time[i] = time[i-1] + dt
    y1[i] = y1[i-1] + v1[i-1]*dt - 0.5*m1*g*dt*dt
    if(y1[i]<0):
        v1[i]= -v1[i-1]
        y1[i] = 0
        flag = True
    else:
        v1[i] = v1[i-1] - m1*g*dt
    y2[i] = y2[i-1] + v2[i-1]*dt - 0.5*m2*g*dt*dt
    if(y2[i]<0):
        y2[i] = 0
        v2[i] = -v2[i-1]
    else:
        v2[i] = v2[i-1] - m2*g*dt

    if y2[i] < y1[i]:
        v1[i] = ((1-r)/(1+r)*v1[i-1])+(2*r)/(1+r)*v2[i-1]
        v2[i] = 2/(1+r)*v1[i-1]+(r-1)/(r+1)*v2[i-1]
        y1[i] = y1[i-1] + v1[i-1]*dt - 0.5*m1*g*dt*dt
        y2[i] = y2[i-1] + v2[i-1]*dt - 0.5*m2*g*dt*dt
    if flag == True:        
        pointv.append(v2[i-1])
        pointy.append(y2[i-1])
        flag = False
        
        
    
    ke1[i] = .5*m1*v1[i]**2
    pe1[i] = m1*g*y1[i]
    ke2[i] = .5*m2*v2[i]**2
    pe2[i] = m2*g*y2[i]
    
    totalEnergy1[i] = ke1[i]+pe1[i]
    totalEnergy2[i] = ke2[i]+pe2[i]
    totalEnergy[i] = ke1[i]+ke2[i]+pe1[i]+pe2[i]
plt.figure()
plt.plot(time,y1)
plt.plot(time,y2)
#plt.plot(time,v1)
plt.xlabel('time')
plt.ylabel('height')
plt.title('bouncing ball')
plt.figure()
plt.scatter(v1,y1)
plt.figure()
plt.scatter(v2,y2)
plt.figure()
plt.scatter(pointv, pointy)
#plt.plot(v2,y2)
```
