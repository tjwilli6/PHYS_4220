```python
#Import Packages
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Momentum Equations 
def momentum(v1i, v2i, r):
    #First Ball Final velcoity 
    v1f = ((1-r)/(1+r))*v1i + ((2*r)/(1+r))*v2i
    #Second Ball Final velocity
    v2f = (2/(1+r))*v1i + ((r-1)/(r+1))*v2i
    
    #if debug:
    #    E0 = v1i**2 + r * v2i**2
    #    E1 = v1f**2 + r *v2f**2
    #    p0 = v1i + r *v2i
    #    p1 = v1f + r * v2f
    
    return v1f, v2f
```

```python
#Energy Equation
def energy(v1,v2,y1,y2,r):
    return(0.5*v1**2 + 0.5*v2**2 * r + y1 + r*y2) / (1+r)
```

```python
# Following mass ratio equations
def ball(r=0, tstop = 10, dt = 0.01, radius = 0.01, y0 = [.5,1], debug = True):
    t = np.arange(0, tstop*np.sqrt(2), dt)
    v1 = np.zeros_like(t)
    y1 = np.zeros_like(t)
    v2 = np.zeros_like(t)
    y2 = np.zeros_like(t)
    
    vball = []
    yball = []
    y1[0] = y0[0]
    y2[0] = y0[1]
    for i in range(1, t.size):
        # If r is equal 
        #if debug:eng = energy(v1[i-1], v2[i-2],y1[i-1],y2[i-1],r)
        v1[i] = v1[i-1] - dt
        y1[i] = y1[i-1] + v1[i] * dt
        
        v2[i] = v2[i-1] - dt
        y2[i] = y2[i-1] + v2[i] *dt
        
        ground = False
        #print(i)
        # mass ratio is between 0 and radius
        if (y1[i] <= radius) and (v1[i] < 0):
            vball.append(v2[i])
            yball.append(y2[i])
            ground = True
            v1[i] = -v1[i-1]
            y1[i] = y1[i-1] + v1[i] *dt
            
        # mass ratio 
        if y2[i] - y1[i] <= radius:
            v1i = v1[i-1]
            v2i = v2[i-1]
            if ground:
                v1i = v1[i]
            v1f, v2f = momentum(v1i, v2i, r)
            v1[i] = v1f
            v2[i] = v2f
            y1[i] = y1[i-1] + v1[i] * dt
            y2[i] = y2[i-1] + v2[i] * dt
            
    return t, v1, v2, y1, y2, vball, yball
```

```python
def graph(r =1, dt = 0.001, radius=0, tstop = 10):
    t,v1,v2,y1,y2,vball,yball = ball(dt=dt, radius=radius, tstop=tstop,r=r)
    plt.subplot(2,1,1)
    plt.plot(t,y1)
    plt.plot(t,y2)
    #plt.ylabel('$\frac{y}{y_0}$')
    plt.gca().xaxis.set_visible(False)
    plt.subplot(2,1,2)
    plt.plot(t,v1)
    plt.plot(t,v2)
    #plt.xlabel('$\frac{t}{t_0}$')
    plt.show()
    #print(v1[:10])
graph(r=.5)
```
