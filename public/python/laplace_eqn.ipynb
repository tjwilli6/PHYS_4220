## 1D example

```python
#Geometry
xmin = 0
xmax = 1
dx = 0.01
x = np.arange(xmin,xmax,dx)

#Boundary conditions
v_x0 = -1
v_x1 = 1

#Initial guess
v_old = np.zeros_like(x)
#Setup boundary conditions
v_old[0] = 1
v_old[-1] = -1
#An array to hold our updated guess
v_new = np.zeros_like(v_old)
v_new[0] = v_old[0]
v_new[-1] = v_old[-1]
n=0
```

```python
for i in range(1,v_old.size-1):
    v_new[i] = 0.5 * (v_new[i-1] + v_old[i+1])
v_old[:] = v_new[:]
n+=1
```

```python
deltaV = np.inf
dV = []
n = 0
while deltaV > 1e-5:
    deltaV = 0
    for i in range(1,v_old.size-1):
        v_new[i] = 0.5 * (v_old[i-1] + v_old[i+1])
        deltaV+=np.abs(v_new[i]-v_old[i])
    deltaV/=v_new.size-2
    dV.append(deltaV)
    v_old[:] = v_new[:]
    n+=1
```

```python
plt.plot(x,v_new)
plt.title(r'{} Iterations: , $\Delta V$: {:.3E}'.format(n,deltaV))
plt.xlabel('x')
plt.ylabel('V');
```

## 2D example

```python
x = np.arange(0,1,0.02)
y = np.arange(0,1,0.02)
xg,yg = np.meshgrid(x,y)
V0 = 1

v_old = np.zeros((x.size,y.size))

v_old[0,:] = V0
v_old[-1,:] = 0

v_new = np.zeros_like(v_old)
v_new[:,:] = v_old[:,:]
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection='3d')
ax.plot_surface(xg.T,yg.T,v_old,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x/a')
ax.set_ylabel('y/a')
#ax.view_init(0,-90)
plt.tight_layout()
```

```python
tstart = time.time()
dV = np.inf
n = 0
while dV > 1e-5:
    for i in range(1,x.size-1):
        for j in range(1,y.size-1):
            v_new[i,j] = (
                v_old[i+1,j] + v_old[i-1,j] +
                v_old[i,j+1] + v_old[i,j-1]
            ) / 4
    dV = np.mean( np.abs(v_new-v_old) )
    v_old[:,:] = v_new[:,:]
    n+=1
tstop = time.time()
print("{:.1f} seconds for {} iterations".format(tstop-tstart,n))
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection='3d')
temp=ax.plot_surface(xg.T,yg.T,v_new,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x/a')
ax.set_ylabel('y/a')
#ax.view_init(0,-90)
plt.tight_layout()
```

## Parallel plate capacitor

```python
dl = 0.025 #grid size
x = np.arange(-1,1,dl)
y = np.arange(-1,1,dl)
s = 0.5 #distance btwn plates
L = 0.5 #length of plates
xg,yg = np.meshgrid(x,y)
#potential of plates
V0 = 1
```

#### Find indices of capacitor plate boundaries

```python
s = 0.5
deltaX = (x.max()-x.min())/2
x_low = deltaX-s/2 #x distance from leftmost boundary
x_hi  = deltaX+s/2
i_left = int(np.ceil(x_low / dl))
i_right = int(np.ceil(x_hi / dl))

print("Left plate located at x={:.2f}, right plate at x={:.2f}".format(x[i_left],x[i_right]))

L = 0.5
deltaY = (y.max()-y.min())/2
y_low = deltaY-L/2
y_hi = deltaY+L/2
j_low = int(np.ceil(y_low/dl))
j_up = int(np.ceil(y_hi/dl))

print("Bottom of plate located at y={:.2f}, top of plate at x={:.2f}".format(y[j_low],y[j_up]))
```

```python
#distance between plates
s = 0.5
i_left = -1
i_right = -1
for i in range(x.size):
    if x[i] > -s/2 and i_left < 0:
        i_left = i
    if x[i] > s/2 and i_right < 0:
        i_right = i
print("Left plate located at x={:.2f}, right plate at x={:.2f}".format(x[i_left],x[i_right]))
```

```python
L = 0.5
j_low = -1
j_up = -1
for j in range(y.size):
    if y[j] > -L/2 and j_low < 0:
        j_low = j
    if y[j] > L/2 and j_up < 0:
        j_up = j
print("Bottom of plate located at y={:.2f}, top of plate at x={:.2f}".format(y[j_low],y[j_up]))
```

```python
v_old = np.zeros((x.size,y.size))

#V=0 on outer box
v_old[0,:] = 0
v_old[-1,:] = 0
v_old[:,0] = 0
v_old[:,-1] = 0

#V=V0 on left capacitor
v_old[i_left,j_low:j_up] = V0
#V=-V0 on right capactior
v_old[i_right,j_low:j_up] = -V0
v_new = np.zeros_like(v_old)
v_new[:,:] = v_old[:,:]
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection='3d')
temp=ax.plot_surface(xg.T,yg.T,v_new,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x/a')
ax.set_ylabel('y/a')
#ax.view_init(0,-90)
plt.tight_layout()
```

```python
tstart = time.time()
dV = np.inf
n = 0
while dV > 1e-5:
    for i in range(1,x.size-1):
        for j in range(1,y.size-1):
            if i in (i_left,i_right):
                if j > j_low and j < j_up:
                    continue
            v_new[i,j] = (
                v_old[i+1,j] + v_new[i-1,j] +
                v_old[i,j+1] + v_new[i,j-1]
            ) / 4
    dV = np.mean( np.abs(v_new-v_old) )
    v_old[:,:] = v_new[:,:]
    n+=1
tstop = time.time()
print("{:.1f} seconds for {} iterations".format(tstop-tstart,n))
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection='3d')
temp=ax.plot_surface(xg.T,yg.T,v_new,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.view_init(0,-90)
plt.tight_layout()
```

```python
Ex = np.zeros_like(v_new)
Ey = np.zeros_like(v_new)

for i in range(1,x.size-1,5):
    for j in range(1,y.size-1,5):
        if i in (i_left,i_right):
            if j > j_low and j < j_up:
                continue
        Ex[i,j] = -(v_new[i+1,j] - v_new[i-1,j]) / (2*dl)
        Ey[i,j] = -(v_new[i,j+1] - v_new[i,j-1]) / (2*dl)
        mag = np.hypot(Ex[i,j],Ey[i,j])
        scale = 100
        arrow = plt.arrow(x[i],y[j],Ex[i,j]/scale,Ey[i,j]/scale,width=0.013,facecolor='green',edgecolor='k')
        plt.gca().add_artist(arrow)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.plot([-s/2,-s/2],[-L/2,L/2],color='red',lw=2)
plt.plot([s/2,s/2],[-L/2,L/2],color='blue',lw=2)
```
